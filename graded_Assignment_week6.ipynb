```python
from transformers import pipeline, set_seed
# Initialize text generation pipeline with distilgpt2
generator = pipeline("text-generation", model="distilgpt2")
set_seed(42)  # for reproducibility

```

```python
# Define the prompt
prompt = "AI is transforming industries by "

# Generate three different continuations
outputs = generator(prompt, max_length=50, num_return_sequences=3, do_sample=True, top_k=50)

# Display results
for i, output in enumerate(outputs, 1):
    print(f"--- Continuation {i} ---")
    print(output['generated_text'])
    print()
```

```python
from transformers import AutoTokenizer

# Load a tokenizer (using distilgpt2 for consistency)
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")

# Input sentence
sentence = "LLMs are powerful tools for natural language understanding."

# Tokenize
tokens = tokenizer.tokenize(sentence)
token_ids = tokenizer.encode(sentence, add_special_tokens=False)
sequence_length = len(token_ids)

# Display results
print("Sentence:", sentence)
print("\nTokens:", tokens)
print("\nToken IDs:", token_ids)
print("\nSequence Length:", sequence_length)
```

```python
# Define three prompts
prompt_summarisation = "Summarise in ≤30 words: AI is revolutionising healthcare, finance, and education by automating tasks, improving predictions, and enabling personalised experiences."
prompt_qa = "Q: Who developed the theory of relativity? A:"
prompt_poem = "Write a 4-line poem about AI."

# Generate outputs
outputs_sum = generator(prompt_summarisation, max_length=40, num_return_sequences=1)
outputs_qa = generator(prompt_qa, max_length=40, num_return_sequences=1)
outputs_poem = generator(prompt_poem, max_length=50, num_return_sequences=1)

# Display results
print("--- Summarisation ---")
print(outputs_sum[0]['generated_text'])
print("\n--- Q&A ---")
print(outputs_qa[0]['generated_text'])
print("\n--- Creative Text Generation ---")
print(outputs_poem[0]['generated_text'])
```

```python
import gensim.downloader as api

# Load GloVe embeddings (50-dimensional)
glove_model = api.load("glove-wiki-gigaword-50")

# Choose words
words = ["king", "queen", "diamond"]

for word in words:
    print(f"\nWord: {word}")
     # First 10 values of the vector
    vector = glove_model[word]
    print("First 10 values of vector:", vector[:10])

    # Top 5 most similar words
    similar = glove_model.most_similar(word, topn=5)
    print("Top 5 similar words:")
    for sim_word, score in similar:
        print(f"  {sim_word}: {score:.4f}")
```

```python
import gensim.downloader as api
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load GloVe embeddings (50-dimensional)
glove_model = api.load("glove-wiki-gigaword-50")

# Define five short sentences
sentences = [
    "AI is changing healthcare rapidly.",
    "Machine learning improves financial predictions.",
    "Jewellery made of diamonds is precious.",
    "Natural language processing enables chatbots.",
    "Emeralds and rubies are valuable gemstones."
]

# Function to compute sentence embedding (average of word vectors)
def sentence_embedding(sentence, model):
    words = [w for w in sentence.lower().split() if w in model]
    if not words:  # handle case where no words are in vocab
        return np.zeros(model.vector_size)
    return np.mean([model[w] for w in words], axis=0)

# Compute embeddings
embeddings = [sentence_embedding(s, glove_model) for s in sentences]

# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)

# Display as a DataFrame for clarity
df = pd.DataFrame(similarity_matrix, index=sentences, columns=sentences)
print(df.round(3))
```

```python
from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Custom input
text = "The new AI-powered tool has significantly improved our team's productivity and reduced errors."

# Run sentiment classification
result = sentiment_analyzer(text)

# Display result
print("Input:", text)
print("Sentiment:", result[0]['label'])
print("Confidence:", round(result[0]['score'], 3))
```
