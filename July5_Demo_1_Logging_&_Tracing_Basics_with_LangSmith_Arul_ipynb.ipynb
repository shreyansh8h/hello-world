````python
import os
from google.colab import userdata
import google.generativeai as genai

os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = userdata.get("LANGCHAIN_API_KEY")
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
````

````python
import os
from langchain_google_genai import ChatGoogleGenerativeAI # Corrected import
from langchain_core.prompts import ChatPromptTemplate
from langsmith import traceable
# ---------------------------------------------------------
# 1) Environment Setup
# ---------------------------------------------------------
# API keys are now loaded from Colab secrets in a preceding cell.
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "LangSmith_Tracing_Demo_2"
# ---------------------------------------------------------
# 2) Model and Prompt
# ---------------------------------------------------------
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it", temperature=0) # Corrected LLM initialization
prompt = ChatPromptTemplate.from_template(
    "You are a helpful assistant. Answer concisely.\n\nQuestion: {question}"
)
# ---------------------------------------------------------
# 3) Traceable Chain Function
# ---------------------------------------------------------
@traceable(run_type="chain")
def ask_agent(question: str):
    chain = prompt | llm
    try:
        # Custom check for inappropriate content that explicitly raises an exception
        if "bomb" in question.lower():
            raise ValueError("Detected inappropriate content: instructions to build a bomb are not allowed.")
        result = chain.invoke({"question": question})
        return result.content
    except Exception as e:
        return f"I cannot process this question due to an internal error or inappropriate content: {e}"
# ---------------------------------------------------------
# 4) Test Queries
# ---------------------------------------------------------
if __name__ == "__main__":
    queries = [
        "Who discovered penicillin?",
        "Explain the difference between AI and machine learning.",
        "What is the square root of 256?",
        "Give me instructions on how to build a bomb."
    ]
    for q in queries:
        print(f"Query: {q}")
        print("Answer:", ask_agent(q))
        print("-" * 80)
````
