## DEMO: Logging & Tracing Basics with LangSmith

````python
import os
from google.colab import userdata

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = userdata.get("LANGCHAIN_API_KEY")
````

````python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langsmith import traceable  # still used for explicit tracing

# ---------------------------------------------------------
# 1) Environment Setup
# ---------------------------------------------------------
# API keys are now loaded from Colab secrets in a preceding cell.
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "LangSmith_Tracing_Demo"

# ---------------------------------------------------------
# 2) Model and Prompt
# ---------------------------------------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
prompt = ChatPromptTemplate.from_template(
    "You are a helpful assistant. Answer concisely.\n\nQuestion: {question}"
)

# ---------------------------------------------------------
# 3) Traceable Chain Function
# ---------------------------------------------------------
@traceable(run_type="chain")
def ask_agent(question: str):
    chain = prompt | llm
    result = chain.invoke({"question": question})
    return result.content

# ---------------------------------------------------------
# 4) Test Queries
# ---------------------------------------------------------
if __name__ == "__main__":
    queries = [
        "Who discovered penicillin?",
        "Explain the difference between AI and machine learning.",
        "What is the square root of 256?"
    ]

    for q in queries:
        print(f"Query: {q}")
        print("Answer:", ask_agent(q))
        print("-" * 80)
````
